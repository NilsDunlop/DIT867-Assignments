```python
#Assginment 2 Data sets with categorical feature 
import pandas as pd 
import numbers as np 
from sklearn.feature_extraction import DictVectorizer
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score 

 
#TASK 1: Working with Dataset with Caterogical features 
#Objective: predict earnings over >50k based on categorical and numeric features

#Step 1: Spliting features from target 
ad_train = pd.read_csv("adult_train.csv")
ad_test = pd.read_csv("adult_test.csv")
ad_train.dropna(), ad_test.dropna()
ad_trainShuffled = ad_train.sample(frac=1, random_state=0)
ad_testShuffled = ad_test.sample(frac=1, random_state=0)

#Split data into preditors and predicted(target)
X_train= ad_trainShuffled.drop('target', axis =1)
X_test =ad_trainShuffled.drop('target', axis=1)
Y_train = ad_trainShuffled['target']
Y_test = ad_testShuffled['target']

#print(Y_test.head())
#print(Y_train.head())


#Step 2: Encoding Features at numbers 
dict_for_X_train = X_train.to_dict('records')
dict_for_X_test = X_test.to_dict('records')
dV = DictVectorizer()
X_train_encoded = dV.fit_transform(dict_for_X_train)
X_test_encoded  = dV.transform(dict_for_X_test)

print(X_test_encoded)
print(X_train_encoded)


#Now measure accruacy of the transformation using DecisionTreesClassifies 
dt_classifer = DecisionTreeClassifier() 
dt_classifer.fit(X_train_encoded,Y_train)
Yguess_dt = dt_classifer.predict(X_test_encoded)
#print(accuracy_score(Y_test, Yguess_dt))
print("This is guessued Y", Yguess_dt)
print("This is actual Y", Y_test)
```
